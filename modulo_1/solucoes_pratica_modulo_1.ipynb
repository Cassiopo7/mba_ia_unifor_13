```python
# ============================================# 2. An√°lise de Despesas# ============================================def analisar_transacoes(transacoes):    '''    Analisa transa√ß√µes financeiras e retorna estat√≠sticas.        Args:        transacoes (list): Lista de dicion√°rios com transa√ß√µes        Returns:        dict: Dicion√°rio com an√°lises realizadas    '''    if not transacoes:        return {}        # Totais por categoria usando dict comprehension    totais_categoria = {        cat: sum(t['valor'] for t in transacoes if t['categoria'] == cat)        for cat in set(t['categoria'] for t in transacoes)    }        # Agrupar por m√™s    meses_gastos = {}    for transacao in transacoes:        data = datetime.strptime(transacao['data'], '%Y-%m-%d')        mes_ano = f"{data.year}-{data.month:02d}"                if mes_ano not in meses_gastos:            meses_gastos[mes_ano] = 0                if transacao['valor'] < 0:  # Apenas gastos            meses_gastos[mes_ano] += abs(transacao['valor'])        # M√™s com maior gasto    mes_maior_gasto = max(meses_gastos.items(), key=lambda x: x[1]) if meses_gastos else None        # M√©dia mensal de gastos    gastos_negativos = [abs(t['valor']) for t in transacoes if t['valor'] < 0]    media_mensal = sum(gastos_negativos) / len(meses_gastos) if meses_gastos else 0        return {        'totais_por_categoria': totais_categoria,        'meses_gastos': meses_gastos,        'mes_maior_gasto': mes_maior_gasto,        'media_mensal_gastos': media_mensal,        'total_receitas': sum(t['valor'] for t in transacoes if t['valor'] > 0),        'total_despesas': abs(sum(t['valor'] for t in transacoes if t['valor'] < 0))    }# Dados de testetransacoes_teste = [    {'data': '2024-01-15', 'descricao': 'Aluguel', 'valor': -1500.00, 'categoria': 'Moradia'},    {'data': '2024-01-20', 'descricao': 'Sal√°rio', 'valor': 5000.00, 'categoria': 'Renda'},    {'data': '2024-02-01', 'descricao': 'Supermercado', 'valor': -350.00, 'categoria': 'Alimenta√ß√£o'},    {'data': '2024-02-05', 'descricao': 'Conta de Luz', 'valor': -120.00, 'categoria': 'Moradia'},    {'data': '2024-02-15', 'descricao': 'Aluguel', 'valor': -1500.00, 'categoria': 'Moradia'},    {'data': '2024-02-20', 'descricao': 'Sal√°rio', 'valor': 5000.00, 'categoria': 'Renda'},    {'data': '2024-02-25', 'descricao': 'Restaurante', 'valor': -180.00, 'categoria': 'Alimenta√ß√£o'},]# Executar an√°liseanalise = analisar_transacoes(transacoes_teste)print("\n" + "=" * 60)print("TESTE 2: An√°lise de Transa√ß√µes")print("=" * 60)print(f"Totais por categoria: {analise['totais_por_categoria']}")print(f"M√™s com maior gasto: {analise['mes_maior_gasto']}")print(f"M√©dia mensal de gastos: R$ {analise['media_mensal_gastos']:.2f}")print(f"Total receitas: R$ {analise['total_receitas']:.2f}")print(f"Total despesas: R$ {analise['total_despesas']:.2f}")
```
